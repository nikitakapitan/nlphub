```python
from datasets import load_dataset
from transformers import pipeline
from nlphub.benchmarks.PerformanceBenchmark import PerformanceBenchmark
from nlphub.vizual import plot_metrics

dataset = load_dataset('clinc_oos', 'plus')

distilbert_ft_clinc_pipe = pipeline("text-classification", model="nikitakapitan/distilbert-base-uncased-finetuned-clinc_oos")

# pb = PerformanceBenchmark(pipeline=distilbert_ft_clinc_pipe, dataset=dataset)
```

```python
from nlphub import ClassificationBenchmark

metric_config = {'accuracy': {},
                 'f1'      : {'average' : 'weighted'}}
pb = ClassificationBenchmark(pipeline=distilbert_ft_clinc_pipe, dataset=dataset['test'], metric_config=metric_config)
```
